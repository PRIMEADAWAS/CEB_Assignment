## 1. Please identify some pros and cons of the EM compare with the K-means
(5 points)

EM (Expectation-Maximization) and K-means are both popular algorithms for clustering data, but they have some significant differences in terms of how they operate and what kind of data they are best suited for. Here are some pros and cons of EM compared to K-means:

Pros of EM:

- Complex distributions: EM can model data with more complex probability distributions than K-means, which assumes that each cluster has a simple Gaussian distribution. EM can model data with mixtures of Gaussians or other complex distributions.

- Handle missing or incomplete data: EM can estimate the missing or incomplete data in the dataset(imputation), which K-means cannot do.

- Different cluster sizes and shapes: EM can model clusters of different sizes and shapes, while K-means is limited to circular or spherical clusters.

- Provides probabilistic cluster assignments: EM assigns probabilities to each point belonging to each cluster, allowing for more nuanced interpretations of the cluster assignments.

Cons of EM:

- More computation: EM requires more computational resources than K-means, which can make it slower and more resource-intensive.

- More data: EM requires a large amount of data to accurately estimate the model parameters, so it may not be suitable for small datasets.

- Converge to local optima: EM may converge to local optima rather than the global optimum, so the results may be sensitive to the initial starting conditions.

Pros of K-means:

- Simple implement: K-means is a simple and intuitive algorithm that is easy to understand and implement.

- Fast: K-means is computationally efficient and can handle large datasets with many variables.

- Well perform with huge, circular or spherical clusters: K-means can perform well with vast, circular or spherical clusters, making it suitable for some types of data.

Cons of K-means:

- Discreate prediction : due to model algorithym there can predict in only single class compare to EM that can predict in probability to each classes.

- Curse of dimensionality problem :  the more number of dimensions increase, the more model efficiency to separate group dataset reduce. 

- Sensitive to initial conditions: K-means is sensitive to the initial conditions, which can result in different clustering solutions.

- Cannot handle missing data: K-means cannot handle missing or incomplete data and requires that all variables be fully assign for each data point.

```python
def impute_em(X, max_iter = 3000,eps = 1e-08):
  # print(X)
  nr,nc = X.shape
  C = np.isnan(X) == False
  # Collect M_iand O_i's
  one_to_nc = np.arange(1,nc + 1, step = 1)
  M = one_to_nc * (C == False) -1
  O = one_to_nc * C -1

  # Generate Mu_0 and Sigma_0
  Mu = np.nanmean(X, axis = 0)
  observed_rows = np.where(np.isnan(sum(X.T)) == False)[0]
  S = np.cov(X[observed_rows, ].T)
  if np.isnan(S).any():
   S = np.diag(np.nanvar(X,axis = 0))
  # Start updating
  Mu_tilde, S_tilde = {}, {}
  X_tilde = X.copy()
  no_conv = True
  iteration = 0

  while no_conv and iteration < max_iter:
    for i in range(nr):
      S_tilde[i] = np.zeros(nc**2).reshape(nc,nc)
      if set(O[i, ]) != set(one_to_nc -1):# missingcomponentexists
        M_i, O_i = M[i, ][M[i, ] != -1], O[i, ][O[i, ] != -1]
        S_MM = S[np.ix_(M_i, M_i)]
        S_MO = S[np.ix_(M_i, O_i)]
        S_OM = S_MO.T
        S_OO = S[np.ix_(O_i, O_i)]
        Mu_tilde[i] = Mu[np.ix_(M_i)] +\
          S_MO @ np.linalg.inv(S_OO) @\
          (X_tilde[i, O_i] - Mu[np.ix_(O_i)])
        X_tilde[i, M_i] = Mu_tilde[i]
        S_MM_O = S_MM - S_MO @ np.linalg.inv(S_OO) @ S_OM
        S_tilde[i][np.ix_(M_i, M_i)] = S_MM_O
    Mu_new = np.mean(X_tilde, axis = 0)
    S_new = np.cov(X_tilde.T,bias = 1) +\
    reduce(np.add, S_tilde.values()) /nr
    no_conv =\
      np.linalg.norm(Mu- Mu_new) >= eps or\
      np.linalg.norm(S - S_new,ord = 2) >= eps
    Mu = Mu_new
    S = S_new
    iteration += 1
    
  result = {
  'mu': Mu,
  'Sigma': S,
  'X_imputed': X_tilde,
  'C': C,
  'iteration': iteration
  }
  return result
```

```python
#We can now impute the missing components of X:
# start = dt.now()
result_imputed = impute_em(X)
# end = dt.now()
#estimation
result_imputed
```
