## a. (4 points) From the below transaction database, generate the frequent itemset and association rules.  
Student#4: use a minimum support threshold = 50%


```python
!pip install mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [["onions", "potatoes", "burger", "cereal"],
["potato chip", "burger","beer", "apple"],
["onions", "potatoes", "potato chip", "burger", "beer"],
["eggs", "onions", "potatoes", "cereal", "potato chip", "burger"],
["potato chip", "beer", "onions"]]
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_food = pd.DataFrame(te_ary, columns=te.columns_)
df_food
```

### $\color{green} {Frequent itemset}$
minimum support threshold = 50%


```python
frequent_itemsetsA = apriori(df_food, min_support=0.5, use_colnames=True)
frequent_itemsetsA
```

### $\color{green} {Association Rules}$
Lift more than 1  
because the lift value is a measure of importance of a rule
if lift > 1 X->Y :  
people who buy item X will buy item Y too

```python
rulesA = association_rules(frequent_itemsetsA, metric="lift", min_threshold=1)
rulesA
```

filter_rules  
filter only rules that have lift more than 1 and confidence value > 0.75 because almost data has 100% or 75% confidence which is lowest then I decide to remove lowest probability

```python
filter_rulesA = rulesA[ (rulesA['lift'] >= 1) & (rulesA['confidence'] > 0.75)]
filter_rulesA
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
url = 'https://raw.githubusercontent.com/PRIMEADAWAS/CEB_Assignment/main/RADI%20608%2C%20602%20Data%20Mining%20and%20Machine%20Learning/Assignment%20Apriori%20%26%20K-Mean%20Clustering/prescriptionDB.csv'
# url = r'/content/drive/MyDrive/Dataset/Assignment/prescriptionDB.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
# replace nan value to '-'
df.fillna('-', inplace=True)
df.head()
```

```python
dataset2 = df.iloc[:,1:].values.tolist()
dataset2[0:10]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset2).transform(dataset2)
df_drug = pd.DataFrame(te_ary, columns=te.columns_)
df_drug
```

```python
# first column '-' should be remove
te.columns_
```

```python
df_drug.drop('-',axis=1,inplace=True)
df_drug
```

```python
def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1
drug_sets = df_drug.applymap(encode_units)
drug_sets
```

### $\color{green} {Frequent itemset}$
minimum support threshold = 1% (support=0.01)


```python
frequent_itemsetsB = apriori(drug_sets, min_support=0.01, use_colnames=True)
frequent_itemsetsB
```

### $\color{green} {Association Rules}$
Lift more than 1  
because the lift value is a measure of importance of a rule
if lift > 1 X->Y :  
people who buy item X will buy item Y too

```python
rulesB = association_rules(frequent_itemsetsB, metric="lift", min_threshold=1)
rulesB
```

minimum confidence=0.50

```python
filter_rulesB = rulesB[ (rulesB['lift'] >= 1) & (rulesB['confidence'] >= 0.5) ]
filter_rulesB
```

## c. (3 points) From prescriptionDB.csv  Perform an Apriorialgorithm to generate the association rules by selecting the top 20 rules at support=0.0001 and the LHS has at least two drugs in the basket.

```python
frequent_itemsetsC = apriori(drug_sets, min_support=0.0001, use_colnames=True)
frequent_itemsetsC['length'] = frequent_itemsetsC['itemsets'].apply(lambda x: len(x))
frequent_itemsetsC.head(10)
```

```python
rulesC = association_rules(frequent_itemsetsC, metric="lift", min_threshold=1)
rulesC['antecedents_length'] = rulesC['antecedents'].apply(lambda x: len(x))
rulesC.head()
```

### Perform an Apriorialgorithm to generate the association rules by selecting the 
top 20 rules at support=0.0001 and the LHS has at least two drugs in the basket. order by confidence

```python
rulesC[(rulesC['antecedents_length'] >= 2) & (rulesC['support'] >= 0.0001)].sort_values(by=['confidence'],ascending=False).head(20)
```
